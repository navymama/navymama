```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
import seaborn as sns
from sklearn.utils import shuffle
Diamonds = sns.load_dataset('diamonds')
```

```python
Diamonds.head()
```

```python
Diamonds.info()
```

```python
def cut_recode(series):
    if series=="Ideal":
        return 0
    if series=="Premium":
        return 1
    if series=="Very Good":
        return 2
    if series=="Good":
        return 3
    if series=="Fair":
        return 4
```

```python
def color_recode(series):
    if series=="D":
        return 0
    if series=="E":
        return 1
    if series=="G":
        return 2
    if series=="H":
        return 3
    if series=="I":
        return 4
    if series=="F":
        return 5
    if series=="J":
        return 6
```

```python
def clarity_recode(series):
    if series=="SI1":
        return 4
    if series=="VS2":
        return 4
    if series=="VS1":
        return 3
    if series=="SI2":
        return 2
    if series=="VVS1":
        return 2
    if series=="SI1":
        return 1
    if series=="IF":
        return 1
    if series=="I1":    
        return 0
    if series=="VVS2":
        return 3
```

```python
Diamonds['cutR']= Diamonds['cut'].apply(cut_recode)
```

```python
Diamonds.head(20)
```

```python
Diamonds['colorR']= Diamonds['color'].apply(color_recode)
```

```python
Diamonds.head(20)
```

```python
Diamonds['clarityR']= Diamonds['clarity'].apply(clarity_recode)
```

```python
Diamonds.head(30)
        
```

```python
Diamonds.head(30)
```

```python
x = Diamonds[['carat', 'cutR','clarityR', 'colorR']]
```

```python
y = Diamonds['price']
```

```python
x_train, y_train, x_test, y_test = train_test_split(x, y,test_size=0.5,random_state=0)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
predictions = lm.predict(y_train)
```
